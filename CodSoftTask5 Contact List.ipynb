```python
contactbook = {}
def add_contact():
    name = input("Enter contact name: ")
    phone = input("Enter phone number: ")
    email = input("Enter email: ")
    address = input("Enter address: ")
    contactbook[name] = {"Phone": phone, "Email": email, "Address": address}
    print("Contact added successfully!")
def view_contacts():
    if not contactbook:
        print("Contact book is empty.")
    else:
        print("Contact List:")
        for name, info in contactbook.items():
            print(f"Name: {name}, Phone: {info['Phone']}")
def search_contact():
    search_term = input("Enter name or phone number to search: ")
    found = False
    for name,info in contactbook.items():
        if name == search_term or info['Phone'] == search_term:
            found = True
            print(f"Name: {name}, Phone: {info['Phone']}, Email: {info['Email']}, Address: {info['Address']}")
    if not found:
        print("Contact not found.")
def update_contact():
    name = input("Enter name of the contact to update: ")
    if name in contactbook:
        print(f"Current Contact Information for {name}:")
        print(contactbook[name])
        phone = input("Enter new phone number (press Enter to keep the current one): ")
        email = input("Enter new email (press Enter to keep the current one): ")
        address = input("Enter new address (press Enter to keep the current one): ")
        if phone:
            contactbook[name]["Phone"] = phone
        if email:
            contactbook[name]["Email"] = email
        if address:
            contactbook[name]["Address"] = address
        print("Contact updated successfully!")
    else:
        print("Contact not found.")
def delete_contact():
    name = input("Enter name of the contact to delete: ")
    if name in contactbook:
        del contactbook[name]
        print("Contact deleted successfully!")
    else:
        print("Contact not found.")
while True:
    print("\nContact Book Menu:")
    print("1. Add Contact")
    print("2. View Contacts")
    print("3. Search Contact")
    print("4. Update Contact")
    print("5. Delete Contact")
    print("6. Exit")
    choice = input("Enter your choice(1,2,3,4,5,6): ")
    if choice == '1':
        add_contact()
    elif choice == '2':
        view_contacts()
    elif choice == '3':
        search_contact()
    elif choice == '4':
        update_contact()
    elif choice == '5':
        delete_contact()
    elif choice == '6':
        print("Thankyou!")
        break
    else:
        print("Invalid choice. Please enter a valid option.")
```
